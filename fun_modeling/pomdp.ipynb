# Partially Observable Markov Processes

This notebook displays plots of picewise linear constraints resulting from solving a POMDP.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("homework")
```

```python
p1 = np.linspace(0, 1)
```

```python
f1 = p1 * 0
```

```python
f2 = 144 * p1 - 72
```

```python
U1 = np.max((f1, f2), axis=0)
```

```python
fig, ax = plt.subplots()
ax.plot(p1, f1, c="b", ls="dashed")
ax.plot(p1, f2, c="b", ls="dashed")
ax.plot(p1, U1, c="b")

ax.text(0.1, 10, r"$a_1$ optimal")
ax.text(.6, 50, r"$a_2$ optimal")

fig.text(0.1, 0.0, r"$s = s_2$")
fig.text(0.8, 0.0, r"$s = s_1$")
ax.set_xlabel(r"$p_1$")
ax.set_ylabel(r"$U_1(b)$")
ax.set_title(r"$U_1(b) = \max_a U_1(b, a)$")
ax.set_ylim(-100, 100);
```

```python
f22 = -4.563 * p1 - 36.288 * (1-p1)
f23 = -33.048 * p1 + 7.776 * (1-p1)
f24 = -37.611 * p1 - 28.512 * (1-p1)
U2 = np.max((f1, f2, f22, f23, f24), axis=0)
```

```python
fig, ax = plt.subplots()
ax.plot(p1, f1, c="b", ls="dashed")
ax.plot(p1, f2, c="b", ls="dashed")
ax.plot(p1, f22, c="b", ls="dashed")
ax.plot(p1, f23, c="b", ls="dashed")
ax.plot(p1, f24, c="b", ls="dashed")

ax.plot(p1, U2, c="b")

ax.text(0.1, 10, r"$a_1$ optimal")
ax.text(.6, 50, r"$a_2$ optimal")

fig.text(0.1, 0.0, r"$s = s_2$")
fig.text(0.8, 0.0, r"$s = s_1$")
ax.set_xlabel(r"$p_1$")
ax.set_ylabel(r"$U_2(b)$")
ax.set_title(r"$U_2(b) = \max_a U_2(b, a)$")
ax.set_ylim(-100, 100);
```
